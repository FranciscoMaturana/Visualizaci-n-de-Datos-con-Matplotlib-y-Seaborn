```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tip = sns.load_dataset('tips')
tip.head(3)
```

### scatterplot

```python
sns.scatterplot(tip, x='total_bill', y = 'tip')
plt.show()
```

```python
sns.scatterplot(tip, x='total_bill', y = 'tip', hue='day')
plt.show()
```

```python
sns.scatterplot(tip, x='total_bill', y = 'tip', hue='day', style='time')
plt.show()
```

```python
sns.scatterplot(tip, x='total_bill', y = 'tip', hue='day', style='time', size = 'size')
plt.legend(loc='center', bbox_to_anchor= (1.12, 0.5))
plt.show()
```

```python
markers = {
    'Lunch': 'D',
    'Dinner': 's'
}
sns.scatterplot(tip, x='total_bill', y = 'tip', hue='day', style='time', size = 'size', markers = markers)
plt.legend(loc='center', bbox_to_anchor= (1.12, 0.5))
plt.show()
```

### lineplot

```python
sns.lineplot(tip, x='total_bill', y = 'tip')

plt.show()
```

```python
sns.lineplot(tip, x='total_bill', y = 'tip', hue='day', style='time', size = 'size')
plt.legend(loc='center', bbox_to_anchor= (1.12, 0.5))
plt.show()
```

### relplot

```python
sns.relplot(tip, x='total_bill', y = 'tip', hue='day', style='time', size = 'size')
#plt.legend(loc='center', bbox_to_anchor= (1.12, 0.5))
plt.show()
```

```python
sns.relplot(tip, x='total_bill', y = 'tip', hue='day', style='time', size = 'size', kind = 'line', col= 'time')
#plt.legend(loc='center', bbox_to_anchor= (1.12, 0.5))
plt.show()
```
