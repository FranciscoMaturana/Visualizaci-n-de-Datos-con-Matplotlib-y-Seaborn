```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

```python
sns.set(style = 'darkgrid')
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

```python
sns.set(style = 'darkgrid', palette = 'ch:s=.25,rot=-.25')
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

```python
sns.set(style = 'darkgrid', palette = 'ch:s=.25,rot=-.25', font="Verdana")
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

```python
sns.set(style = 'darkgrid', palette = 'ch:s=.25,rot=-.25', font_scale = 2)
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```
