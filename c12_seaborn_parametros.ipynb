```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tip = sns.load_dataset('tips')
tip
```

```python
sns.displot(data=tip, x='total_bill')
plt.show()
```

```python
sns.displot(data=tip, x='total_bill', y = 'tip')
plt.show()
```

```python
sns.displot(data=tip, x='total_bill', y = 'tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tip, x='total_bill', hue='sex')
plt.show()
```

```python
sns.displot(data=tip, x='total_bill', hue='sex', kind = 'kde')
plt.show()
```

```python
sns.displot(data=tip, x='total_bill', hue='sex', kind = 'kde', legend=False)
plt.show()
```

```python
sns.displot(data=tip, x='total_bill', hue='sex', kind = 'kde', palette='light:#8A9')
plt.show()
```
