```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
```

```python
tip = sns.load_dataset('tips')
tip
```

### Hisplot

```python
sns.histplot(data=tip, x='tip' )
plt.show()
```

```python
sns.histplot(data=tip, x='tip', bins= 10 )
plt.show()
```

```python
sns.histplot(data=tip, x='tip', cumulative = True )
plt.show()
```

```python
sns.histplot(data=tip, x='tip', hue='sex' )
plt.show()
```

```python
sns.histplot(data=tip, x='tip', stat = 'frequency' )
plt.show()
```

```python
sns.histplot(data=tip, x='tip', stat = 'probability' )
plt.show()
```

```python
sns.histplot(data=tip, x='tip', stat = 'percent' )
plt.show()
```

```python
sns.histplot(data=tip, x='tip', stat = 'density' )
plt.show()
```

```python
sns.histplot(data=tip, x='tip', hue='sex', multiple = 'stack' )
plt.show()
```

```python
sns.histplot(data=tip, x='tip', hue='sex', multiple = 'layer' )
plt.show()
```

```python
sns.histplot(data=tip, x='tip', hue='sex', multiple = 'dodge' )
plt.show()
```

```python
sns.histplot(data=tip, x='tip', hue='sex', multiple = 'fill' )
plt.show()
```

### kdeplot

```python
sns.kdeplot(data=tip, x='tip', hue = 'sex')
plt.show()
```

```python
sns.kdeplot(data=tip, x='tip', hue = 'sex', cumulative = True)
plt.show()
```

```python
sns.kdeplot(data=tip, x='tip', hue = 'sex', fill = True)
plt.show()
```

```python
sns.kdeplot(data=tip, x='tip', hue = 'sex', bw_adjust = 100)
plt.show()
```

### ecdfplot

```python
sns.ecdfplot(data=tip,x='tip', hue ='sex')
plt.show()
```

```python
sns.ecdfplot(data=tip,x='tip', hue ='sex', stat = 'count')
plt.show()
```

### Displot

```python
sns.displot(data=tip,x='tip', hue ='sex')
plt.show()
```

```python
sns.displot(data=tip,x='tip', hue ='sex', kind = 'kde')
plt.show()
```

```python
sns.displot(data=tip,x='tip', hue ='sex', kind = 'hist')
plt.show()
```

```python
sns.displot(data=tip,x='tip', hue ='sex', kind = 'kde', multiple = 'stack')
plt.show()
```
