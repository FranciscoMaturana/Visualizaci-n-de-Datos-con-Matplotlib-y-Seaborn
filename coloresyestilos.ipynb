```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5,5,21)
```

```python
print(plt.style.available)
```

```python
plt.style.use('ggplot')
```

### Matlab

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x,'r--')
ax.plot(x,x+1,'b-')
ax.plot(x,x+2,'y.-')
ax.plot(x,x+3,'go:')
```

### Pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x, color = 'green', alpha = 0.3, linewidth = 5, linestyle='-', marker = 'o', markersize = 15, markerfacecolor ='red')
ax.plot(x,x+1, color = 'blue', linewidth = 20, linestyle='--', marker = 'x')
ax.plot(x,x+2, color = '#66FF66', linewidth = 3, linestyle='dashed', marker = 'v')
ax.plot(x,x+3, color = '#CCFFCC', linewidth = 1, linestyle=':', marker = 'P')
```
