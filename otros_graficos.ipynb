```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,5000,100)
data
```

```python
plt.style.use('ggplot')
```

### Histograma

```python
plt.hist(data)
plt.show()
```

# bins

```python
plt.hist(data, bins=50)
plt.show()
```

# width

```python
plt.hist(data, bins=50, width=50)
plt.show()
```

# histtype

```python
plt.hist(data, bins=50, histtype='bar')
plt.show()
```

```python
plt.hist(data, bins=50, histtype='step')
plt.show()
```

### Boxplot

```python
plt.boxplot(data)
plt.show()
```

# vert

```python
plt.boxplot(data, vert=False)
plt.show()
```

# patch_artist

```python
plt.boxplot(data, vert=False, patch_artist=True)
plt.show()
```

# notch

```python

plt.boxplot(data, vert=False, notch = True)
plt.show()
```

```python
#Agregando data anomala
data = np.append(data, 15000)
data
```

```python
plt.boxplot(data, vert=False, notch = True)
plt.show()
```

```python
plt.boxplot(data, vert=False, notch = True, showfliers=False)
plt.show()
```

### Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2
colors = np.random.rand(N)
```

```python
x
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y,s=area)
plt.show()
```

```python
plt.scatter(x,y,s=area, c = colors)
plt.show()
```

```python
plt.scatter(x,y,s=area, c = colors, marker = 'x')
plt.show()
```

```python
plt.scatter(x,y,s=area, c = colors, marker = 'o', alpha = 0.5)
plt.show()
```
