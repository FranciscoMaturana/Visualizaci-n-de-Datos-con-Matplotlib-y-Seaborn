```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tip = sns.load_dataset('tips')
tip.head(3)
```

### countplot

```python
sns.countplot(tip,x = 'day', hue='sex')
plt.show()
```

```python
sns.countplot(tip,y = 'day', hue='sex')
plt.show()
```

### stripplot

```python
sns.stripplot(tip,x = 'day',y='total_bill', hue='sex')
plt.show()
```

```python
plt.figure(figsize = (8,8))
sns.stripplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True)
plt.show()
```

### swarmplot

```python
plt.figure(figsize = (8,8))
sns.swarmplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True)
plt.show()
```

### boxplot

```python
plt.figure(figsize = (8,8))
sns.boxplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True)
plt.show()
```

```python
plt.figure(figsize = (8,8))
sns.boxplot(tip,x='total_bill')
plt.show()
```

### boxplot + swarnplot

```python
plt.figure(figsize = (8,8))
sns.boxplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True)
sns.swarmplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True, palette = 'ch:s=.25,rot=-.25')
plt.show()
```

```python
plt.figure(figsize = (8,8))
sns.boxplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True)
sns.swarmplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True, palette = 'ch:s=.25,rot=-.25', marker = '<')
plt.show()
```

### violinplot

```python
plt.figure(figsize = (8,8))
sns.violinplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True, palette = 'ch:s=.25,rot=-.25')

plt.show()
```

```python
plt.figure(figsize = (8,8))
sns.violinplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True, split = True)

plt.show()
```

```python
sns.violinplot(tip,x ='total_bill')
```

### catplot

```python
sns.catplot(tip,x = 'day',y='total_bill', hue='sex')
plt.show()
```

```python
sns.catplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True)
plt.show()
```

```python
sns.catplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True, kind='swarm')
plt.show()
```

```python
sns.catplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True, kind='box')
plt.show()
```

```python
sns.catplot(tip,x = 'day',y='total_bill', hue='sex', dodge = True, kind='box', col= 'time')
plt.show()
```
