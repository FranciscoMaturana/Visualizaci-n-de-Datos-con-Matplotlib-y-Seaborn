```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tip = sns.load_dataset('tips')
tip.head(3)
```

```python
tip.corr()
```

### heatmap

```python
sns.heatmap(tip.corr())
plt.show()
```

```python
sns.heatmap(tip.corr(), annot=True)
plt.show()
```

```python
sns.heatmap(tip.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.heatmap(tip.corr(), annot=True, cmap='coolwarm', linewidths= 5)
plt.show()
```

```python
sns.heatmap(tip.corr(), annot=True, cmap='coolwarm', linewidths= 5, linecolor='lightblue')
plt.show()
```

```python
sns.heatmap(tip.corr(), annot=True, cmap='coolwarm', linewidths= 5, linecolor='lightblue',
            vmin = 0, vmax = 1
            )
plt.show()
```

```python
sns.heatmap(tip.corr(), annot=True, cmap='coolwarm', linewidths= 5, linecolor='lightblue',
            vmin = 0, vmax = 1, cbar=False
            )
plt.show()
```
