```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-100,100,11)
y = x**2+5*x + 2
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r')
plt.show()
```

```python
plt.plot(x,y,'rx')
plt.show()
```

```python
plt.plot(x,y,'rx-')
plt.show()
```

```python
plt.plot(x,y,'yD:')
plt.show()
```

```python
plt.hist(x)
plt.show()
```

```python
plt.pie(x[x > 0])
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.boxplot(x)
plt.show()
```
