```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['India', 'Japan', 'Mexico', 'Colombia', 'Germany']
population = [1000, 800, 900, 1000, 300]
```

```python
plt.style.use('ggplot')
```

### Width

```python
plt.bar(country, population, width = 0.5)
```

```python
plt.bar(country, population, width = 0.1)
```

### Color

```python
plt.bar(country, population, width = 0.5, color = ['b'])
```

```python
plt.bar(country, population, width = 0.5, color = ['b','red'])
```

### xticks

```python
plt.bar(country, population, width = 0.5, color = ['b','red'])
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'))
plt.show()
```

### rotation

```python
plt.bar(country, population, width = 0.5, color = ['b','red'])
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation = 45)
plt.show()
```

```python
plt.barh(country, population)
```
