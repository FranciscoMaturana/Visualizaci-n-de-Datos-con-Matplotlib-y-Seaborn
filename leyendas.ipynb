```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5,5,21)
y = np.sin(x)
```

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x,y,'ro-',label='$sin(x)$')
axes[0].set_title('y = sin(x)')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x,'b--')
axes[1].set_title('relacin Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
plt.plot(x,y,'bo--', label = '$sin(x)$')
plt.title('ttulo')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(loc='lower right')
```
